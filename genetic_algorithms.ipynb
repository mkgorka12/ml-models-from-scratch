## OneMax problem

```python
from src.util.genetic_algs_utils import onemax_create_population
from src.util.fitnesses import onemax_fitness
from src.util.selection import onemax_select_best
from src.util.crossing import onemax_single_point_crossing
from src.util.mutations import onemax_mutation
```

```python
MUTATION_PROB = 0.5

genetic_alg = GeneticAlgorithm(
    create_population_func=onemax_create_population,
    fitness_func=onemax_fitness,
    selection_func=onemax_select_best,
    crossover_func=onemax_single_point_crossing,
    mutation_func=onemax_mutation,
    mutation_probability=MUTATION_PROB
)
```

```python
POPULATION_SIZE = 10

genotypes, best = genetic_alg.run(POPULATION_SIZE, POPULATION_SIZE, verbose=True)
print('Genotypes:', genotypes, 'Best:', best)
```
