# Decision tree

```python
import pandas as pd
import numpy as np

from src.util.decision_tree_utils import fill_dataframe_random_bools
from src.models.decision_tree import DecisionTreeBuilder
```

```python
rng = np.random.default_rng(0)
df = pd.DataFrame(
    columns=[
        "Chest pain",
        "Blood circulation normal",
        "Arterial blockade",
        "Heart attack",
    ]
)

df = fill_dataframe_random_bools(rng, df, 4)
display(df.describe())
display(df.head())
```

```python
target = "Heart attack"
dtb = DecisionTreeBuilder(target)
dt = dtb.build(df)
```

```python
patient = {
    "Chest pain": True,
    "Blood circulation normal": True,
    "Arterial blockade": False,
}
patient_series = pd.Series(data=patient, index=patient.keys())
display(patient_series)
```

```python
pred = dt.predict(patient_series, target)
```
